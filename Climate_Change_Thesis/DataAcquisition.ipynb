```python
import requests
import os

# Download predictor files:
# ========================
# Default: Current location of this script
dest = os.getcwd()
start_x = 126
end_x = 128
start_y = 43
end_y = 50
model_selection = ["MPI-ESM1-2-HR_historical", "MPI-ESM1-2-HR_ssp126", "MPI-ESM1-2-HR_ssp370", "MPI-ESM1-2-HR_ssp585"]

for model in model_selection:
    for x_i in range(start_x, end_x + 1):
        for y_i in range(start_y, end_y + 1):
            url = f"https://climate-scenarios.canada.ca/files/CMIP6_predictors/{model}/{model}_BOX_{str(x_i).zfill(3)}X_{str(y_i).zfill(2)}Y.zip"
            print(url)  # Prints URL that will be downloaded
            file_name = url.split('/')[-1]
            print(f"{dest}/{file_name}")  # Prints filename that will be downloaded
            
            # Ensure the destination directory exists
            os.makedirs(dest, exist_ok=True)
            
            # Download the file
            response = requests.get(url)
            with open(os.path.join(dest, file_name), 'wb') as file:
                file.write(response.content)
```

```python
import re
import zipfile

# Download predictor files:
# ========================
# Default: Current location of this script
dest = os.getcwd()
start_x = 1
end_x = 16
start_y = 43
end_y = 50
model_selection = ["MPI-ESM1-2-HR_historical", "MPI-ESM1-2-HR_ssp126", "MPI-ESM1-2-HR_ssp370", "MPI-ESM1-2-HR_ssp585"]

for model in model_selection:
    for x_i in range(start_x, end_x + 1):
        for y_i in range(start_y, end_y + 1):
            url = f"https://climate-scenarios.canada.ca/files/CMIP6_predictors/{model}/{model}_BOX_{str(x_i).zfill(3)}X_{str(y_i).zfill(2)}Y.zip"
            print(url)  # Prints URL that will be downloaded
            file_name = url.split('/')[-1]
            print(f"{dest}/{file_name}")  # Prints filename that will be downloaded
            
            # Ensure the destination directory exists
            os.makedirs(dest, exist_ok=True)
            
            # Download the file
            response = requests.get(url)
            with open(os.path.join(dest, file_name), 'wb') as file:
                file.write(response.content)


# Download predictor files:
# ========================
# Default: Current location of this script
dest = os.getcwd()  # Sets the destination to the current working directory
start_x = 126
end_x = 16
start_y = 43
end_y = 50
model_selection = ["MPI-ESM1-2-HR_historical", "MPI-ESM1-2-HR_ssp126", "MPI-ESM1-2-HR_ssp370", "MPI-ESM1-2-HR_ssp585"]

for model in model_selection:
    for x_i in range(start_x, end_x + 1):
        for y_i in range(start_y, end_y + 1):
            url = f"https://climate-scenarios.canada.ca/files/CMIP6_predictors/{model}/{model}_BOX_{str(x_i).zfill(3)}X_{str(y_i).zfill(2)}Y.zip"
            print(url)  # Prints URL that will be downloaded
            file_name = url.split('/')[-1]
            print(f"{dest}/{file_name}")  # Prints filename that will be downloaded
            
            # Ensure the destination directory exists
            os.makedirs(dest, exist_ok=True)
            
            # Download the file
            response = requests.get(url)
            with open(os.path.join(dest, file_name), 'wb') as file:
                file.write(response.content)

# Zips the folder of predictor data (optional)
zip_filename = os.path.join(dest, "Predictor.zip")
regex_pattern = r"^(CanESM5|ECMWF_ERA5|MPI-ESM1-2-HR|NCEP-DOE|NorESM2-MM)_(historical|ssp119|ssp126|ssp245|ssp370|ssp585)?_?BOX_[0-9]{3}X_[0-9]{2}Y\.zip$"

with zipfile.ZipFile(zip_filename, 'w') as zipf:
    for foldername, subfolders, filenames in os.walk(dest):
        for filename in filenames:
            if re.match(regex_pattern, filename):
                zipf.write(os.path.join(foldername, filename), filename)
                os.remove(os.path.join(foldername, filename))  # Remove the file after adding it to the zip
```
